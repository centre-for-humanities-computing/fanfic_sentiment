```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('../data/merged_sentiment_data_9000_fics.csv')

# add 9000fics_ as a prefix to the textid 
df1['text_id'] = '9kfics_' + df1['text_id'].astype(str)

# drop the "published.1" column
df1 = df1.drop(columns=['published.1'])

df1.head()
```

```python
df2 = pd.read_csv('../data/merged_sentiment_data_MythFic_texts.csv')

# add fandom_label column
df2['fandom_label'] = "AGRL"

# add mythfics as a prefix to the textid 
df2['text_id'] = 'mythfics_' + df2['text_id'].astype(str)

df2.head()
```

```python
df_merged = pd.concat([df1, df2], ignore_index=True)

# add a wordcount column
df_merged['wordcount'] = df_merged['text'].apply(lambda x: len(x.split()))

df_merged.head()
```

```python
print(df_merged['sentences'].apply(type).value_counts())
import ast

df_merged['sentences'] = df_merged['sentences'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)
```

```python
# get the counts
# and all that per fandom
# sentences per fandom
sentences_per_fandom = df_merged.groupby('fandom_label')['sentences'].apply(lambda x: sum(len(s) for s in x if isinstance(s, list))).reset_index()
sentences_per_fandom.columns = ['fandom', 'total_sentences']
print("Total sentences per fandom:")
print(sentences_per_fandom)

# words per fandom
words_per_fandom = df_merged.groupby('fandom_label')['text'].apply(lambda x: sum(len(text.split()) for text in x if isinstance(text, str))).reset_index()
words_per_fandom.columns = ['fandom', 'total_words']
print("Total words per fandom:")
print(words_per_fandom)

# stories per fandom
stories_per_fandom = df_merged.groupby('fandom_label').size().reset_index(name='total_stories')
print("Total stories per fandom:")
print(stories_per_fandom)
```

```python
# get overall sentences, words, and stories
overall_sentences = sentences_per_fandom['total_sentences'].sum()
overall_words = words_per_fandom['total_words'].sum()
overall_stories = stories_per_fandom['total_stories'].sum()
print(f"Overall sentences: {overall_sentences}")
print(f"Overall words: {overall_words}")
print(f"Overall stories: {overall_stories}")
```
