```python
import pandas as pd
```

```python
df1 = pd.read_csv('../data/merged_sentiment_data_9000_fics.csv')

# add 9000fics_ as a prefix to the textid 
df1['text_id'] = '9kfics_' + df1['text_id'].astype(str)

# drop the "published.1" column
df1 = df1.drop(columns=['published.1'])

df1.head()
```

```python
df2 = pd.read_csv('../data/merged_sentiment_data_.csv')

# add fandom_label column
df2['fandom_label'] = "AGRL"

# add mythfics as a prefix to the textid 
df2['text_id'] = 'mythfics_' + df2['text_id'].astype(str)

df2.head()
```

```python
df_merged = pd.concat([df1, df2], ignore_index=True)
df_merged.head()
```
